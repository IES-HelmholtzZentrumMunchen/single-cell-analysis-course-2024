### StarDist
Luckily, several very good neural networks have already been implemented that work very well for standard fluorescent microscopy images, but can also be 're-trained' for your specific data sets. The whole process of training the network and a comprehensive comparison of currently used and developed architectures would go beyond the scope of this course; but we will apply one approach that has been nicely implemented in Python. For the details, navigate to the [StarDist github repository](https://github.com/mpicbg-csbd/stardist) and have a look at the 2018 research paper [Cell Detection with Star-convex Polygons](https://arxiv.org/abs/1806.03535) by Uwe Schmidt, Martin Weigert, Coleman Broaddus, and Gene Myers.

The following is an adaptation of their [example notebook](https://github.com/mpicbg-csbd/stardist/blob/master/examples/2D/3_prediction.ipynb) for a 2D segmentation pipeline.

```python
from __future__ import print_function, unicode_literals, absolute_import, division
import sys
import numpy as np
import matplotlib
matplotlib.rcParams["image.interpolation"] = None
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

from glob import glob
from tifffile import imread
from csbdeep.utils import Path, normalize
from csbdeep.io import save_tiff_imagej_compatible

from stardist import random_label_cmap, _draw_polygons, export_imagej_rois
from stardist.models import StarDist2D

np.random.seed(6)
lbl_cmap = random_label_cmap()
```

```python
X = sorted(glob('../evaluation_data/images/*.tiff'))
X = list(map(imread,X))

n_channel = 1 if X[0].ndim == 2 else X[0].shape[-1]
axis_norm = (0,1)   # normalize channels independently
# axis_norm = (0,1,2) # normalize channels jointly
```

Let's load the models underlying the network; the first one is the pre-trained model from the original paper, the second is our own model, trained on approx. 100 images.

```python
model1 = StarDist2D.from_pretrained('2D_demo')
model2 = StarDist2D(None, basedir='../nn_models/', name='dragonfly-40x-2x')
```

Importantly, the models were trained on ***normalized*** images. This means that the image intensities have been normalized such that the resulting image has values between 0 and 1. The Normalization suggested by the StarDist authors also clips off really dark and really bright pixels (< 1% and > 99.8 %). This helps the network to generealize better and not, for example, learn to classify only really bright pixels as cells that may be image artifacts.

```python
img = normalize(X[0], 1,99.8, axis=axis_norm)
labels_1, details_1 = model1.predict_instances(img) # model of the paper
labels_2, details_2 = model2.predict_instances(img) # our own model
```

This was it! Now we can compare both models.

```python
fig, ax = plt.subplots(1, 2, figsize=(20,10))
ax[0].imshow(img, clim=(0,1), cmap='gray')
ax[0].imshow(labels_1, cmap=lbl_cmap, alpha=0.5)
ax[0].axis('off')
ax[1].imshow(img, clim=(0,1), cmap='gray')
ax[1].imshow(labels_2, cmap=lbl_cmap, alpha=0.5)
ax[1].axis('off')
plt.tight_layout();
```

Hopefully you can appreciate that our own model performs much better! One reason is that the original model was trained on a data set from lower magnification and resolution images and probably also a different cell type. Let's test the performance by calculating the IOU score for the results.

```python
test_mask_files = sorted(glob('../evaluation_data/masks/*.tiff'))
test_mask = imread(test_mask_files[0])
```

```python
for StarDist_label, Model_name in zip([labels_1, labels_2],["pre-trained 2D model", "Dragonfly 40x + 2.0x model"]):
    result1 = StarDist_label > 0
    result2 = test_mask > 0
    intersection = np.logical_and(result1, result2)
    union = np.logical_or(result1, result2)
    iou_score = np.sum(intersection) / np.sum(union)
    print("IOU score of the {} is {:4f}".format(Model_name, iou_score))
```

Luckily for us, the effort of hand-labeling our own data payed off and we get an IUO score of 93.4%! This is very good for image segmentation and classification.
