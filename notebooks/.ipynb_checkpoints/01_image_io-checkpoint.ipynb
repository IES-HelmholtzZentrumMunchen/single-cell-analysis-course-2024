# Image input, output and display

First, we import the required Python modules:
  * `imread` for reading images.
  * Remember, images are just matrices of numbers, so `numpy` comes in handy. Python programmes usually use the alias `np`
  * `matplotlib.pyplot` provides plotting functions. Python programmes usually ues the alias `plt`
  * `glob` and `os` to access our file system.

```python
from skimage.io import imread
import numpy as np
from matplotlib import pyplot as plt
from glob import glob
import os
```

`%` indicates a Jupyter "magic" function. Here we set up plotting within the notebook.
`ipywidgets` provides an interactive and fun way to change variables and see the result right away. We will play around with it further below to get an intuition for how changing image minimum and maximum display affects the brightness and contrast.

```python
%matplotlib inline
from ipywidgets import interact, interactive, fixed
```

## 1. Find image files

We have different ways to find image files.
`glob` provides a command-line like behavior to filter files.

```python
image_files = sorted(glob('../sample_data/*.tiff'))
```

`os.walk` descends into the directory and outputs all files and subfolders. We can get all tiff files by using list comprehension.

```python
image_files = [os.path.join(root, file) for root, path, files in os.walk('../sample_data/') for file in files ]
```

Either method generates a list of image files;

```python
image_files
```

## 2. Read in images

Let's just read in the very first file. Remember, Python indexing starts at 0!

```python
image = imread(image_files[0])
```

What are the image dimensions (X, Y, channels, z-slices, time points)?

```python
image.shape
```

Let's copy the first channel into a new array.

```python
channel_0 = image[:,:,0]
```

What's inside? Remember, images are just arrays (in Python, these are numpy arrays).

```python
channel_0
```

## 3. Explore image data

What if we want to know some basic properties of our image?

Try out the following to get the image **maximum**, **minimum**, and **mean** value:

```python
np.max(channel_0), np.min(channel_0), np.mean(channel_0)
```

Sometimes, the histogram is useful. Try to understand what it shows. You see plotted the number a certain intensity value (more specifically, the range fitting in one bin) occurs on Y plotted against the value on X.

```python
_  = plt.hist(np.ravel(channel_0), bins=50)
```

## 4. Display image

How to have a look at this single image? Pyplot can take care of this:

```python
plt.figure(figsize=(10,10))
plt.imshow(channel_0,cmap='gray',vmin=300,vmax=4000);
```

Because microscopy images rarley use the whole dynamic range of the image evenly, we have to adjust the minimum and maximum displayed valus with `vmin` and `vmax`. All scientific images are in gray scale, but we can show them in false colors to get a better impression how intensities are distributed. Play around with the following script, the sliders adjust minimum and maximum display values. You can even change the colors and select other channels. Note that here I set the maximum available value to the maximum intensity of the image. However, the example images are saved in 16 bit, so you have 65536 possible values. But if you set the maximum to 65535, the image would most likely appear black on the screen.

```python
def interactive_plot(img,ch=0,low=0,high=np.max(image),col_map='gray'):
    fig, ax = plt.subplots(1,1,figsize=(10,10))
    channel = img[:,:,ch]
    ax.imshow(channel, vmin=low, vmax=high, cmap=col_map)
    ax.axis('off')
    plt.show()
    
p = interactive(interactive_plot, {'manual': 'true'}, ch=(0,2), low=(0,np.max(image)), high=(0,np.max(image)), img=fixed(image), col_map=['gray','jet','viridis','plasma'])
p
```
